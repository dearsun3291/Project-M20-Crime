# Visualization

Univeriate Analysis

```python
# Checking data distribution
for i in cr.describe().columns:
    sns.distplot(cr[i])
    plt.show()
```

```python
list = ['Area_Name', 'Year', 'Group_Name', 'Sub_Group_Name']
for i in list:
    print(cr[i].value_counts())
    plt.figure(figsize = (15,5))
    sns.countplot(cr[i])
    plt.xticks(rotation = 90)
    plt.show()
```

Bi-variate Analysis

```python
sns.pairplot(cr)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(cr.Year, cr.Area_Name, hue = cr.Sub_Group_Name, cmap = 'viridis')
plt.xticks(rotation=90)
plt.show()
```
